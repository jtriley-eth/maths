# Ring Theory

ring theory studies the algebraic structures of rings

## Rings

### Ring

a ring is a set equipped with two binary operators, commonly referred to as
addition and multiplication such that addition forms an abelian group with the
set, multiplication forms a monoid with the set, and addition and multiplication
are connected through distributivity

reminder that an abelian group requires that the operator exhibit associativity,
an identity element, inverse element, and commutativity while a monoid only
requires associativity and an identity element.

> while these operators are commonly referred to as addition and multiplication,
> it can be any two operators exhibiting the necessary structures; so for
> abstraction's sake, we represent these operations as $\star$ and $\diamond$
> respectively.

$$
\begin{gather*}
    \forall a, b \in S : a \star b = c \in S \\
    \forall a, b, c \in S : (a \star b) \star c = a \star (b \star c) \\
    \forall a \in S, \exists e \in S : a \star e = e \star a = a \\
    \forall a \in S, \exists a^{-1} \in S : a \star a^{-1} = a^{-1} \star a = e \\
    \forall a, b \in S : a \star b = b \star a \\
    \\
    \forall a, b, c \in S : (a \diamond b) \diamond c = a \diamond (b \diamond c) \\
    \forall a \in S, \exists e \in S : a \diamond e = e \diamond a = a \\
    \\
    \forall a, b, c \in S : a \diamond (b \star c) = (a \diamond b) \star (a \diamond c)
\end{gather*}
$$

### Rig

a rig is a ring without the requirement of the existence of an additive inverse

### Commutative Ring

a commutative ring is a ring such that multiplication is also commutative:

$$\forall a, b \in S : a \diamond b = b \diamond a$$

a nonzero commutative ring is a commutative ring such that at least one element
is not the zero element; aka

$$R \neq { 0 }$$

### Integral Domain

an integral domain is a nonzero commutative ring with no zero divisors. that is
to say that for any two elements, if the product of those two elements is zero,
then one of the two elements MUST be zero:

$$\forall a, b \in S : a \diamond b = 0 \implies a = 0 \lor b = 0$$

### Division Ring

division rings aka skewfields are equipped with a multiplicative inverse where
the multiplicative identity element is not zero, this corresponds to division

$$\forall a \in S, \exists e^{-1} : a \diamond a^{-1} = a^{-1} $$

### Field

a field is a commutative division ring

$$\forall a, b \in S : a \diamond b = b \diamond a$$
$$\forall a \in S, \exists e^{-1} : a \diamond a^{-1} = a^{-1} $$

### Field Extension

if a field $K$ is a subfield of another field $L$ aka $K \subseteq L$, then the
$L$ field is an extension field of $K$

an example of an extension field is the complex numbers $\mathbb{C}$ which extend the
real numbers $\R$:

$$\R \subseteq \mathbb{C}$$

### Galois Field

a galois field is a field with a finite number of elements; the most common
being a set of integers modulo a prime number

### Subring

a subring is a subset of a ring's set such that the subring forms a ring with
the same operators and identities

$$S \subseteq R : (R, \star, \diamond, 0, 1) \rightarrow (S, \star, \diamond, 0, 1)$$

### Ideal

an ideal is a subring with absorbtion properties:

a left ideal is a subring $I \subseteq S$ where the additive inverse is in $I$,
the sum of any two elements in $I$ is also in $I$, and the product of two
elements of $I$ is an element of $I$ if _at least_ the left operand is in $I$

$$
\begin{gather*}
I \subseteq S \\
0 \in I \\
x \star y \in I \implies x, y \in I \\
x \diamond y \in I \implies x \in I
\end{gather*}
$$

a right ideal is a subring $I \subseteq S$ where the additive inverse is in $I$,
the sum of any two elements in $I$ is also in $I$, and the product of two
elements of $I$ is an element of $I$ if _at least_ the right operand is in $I$

$$
\begin{gather*}
I \subseteq S \\
0 \in I \\
x \star y \in I \implies x, y \in I \\
x \diamond y \in I \implies y \in I
\end{gather*}
$$

a two-sided ideal is both a left and right ideal:

$$
\begin{gather*}
I \subseteq S \\
0 \in I \\
x \star y \in I \implies x, y \in I \\
x \diamond y \in I \implies x \in I \lor y \in I
\end{gather*}
$$

### Quotient Ring

a quotient ring $R/I$ is a ring whose set is a set of subsets aka partitions
of the set $R$, partitioned by an ideal $I$

> definition of this one is a bit tricky, so here's an example

an example would be the ring of integers:

$$R = (\Z, +, *, 0, 1)$$

with an ideal such that all elements of the ideal's set are integers that are
multiples of $3$ (note the number $3$ is arbitrary). we define the set of
integers that are multiples of $3$ as $3\Z$

$$3\Z = \{ x | x \in \Z \land x \pmod{3} \equiv 0 \}$$
$$I = (3\Z, +, *, 0, 1)$$

the quotient ring $R/I$ is a quotient set such that the elements of the subsets
are defined as follows:

$$[0] = 0 + 3\Z = \{ .., -6, -3, 0, 3, 6, .. \}$$
$$[1] = 1 + 3\Z = \{ .., -5, -2, 1, 4, 7, .. \}$$
$$[2] = 2 + 3\Z = \{ .., -4, -1, 2, 5, 8, .. \}$$
$$R/I = \{ [0], [1], [2] \}

### Polynomial Ring

a polynomial ring, denoted $R[X]$, is a commutative ring with coefficients in
$R$ and variables aka indeterminates in $X$ with a function
$f : X \rightarrow R[X]$. a polynomial is an element of a polynomial ring.

#### Polynomials

a polynomial $p$, being an element of a polynomial ring $R[X]$, generally takes
the form as follows:

$$a_n \in R, x \in X : p \in R[X] = \sum_{n=0}^n a_n x^{n}$$

or more simply written:

$$(a_0 x^0 + a_1 x^1 + a_2 x^2 + .. + a_n x^n)  \in R[X]$$

univariate polynomials have a single variable $x \in X$

multivariate polynomials have multiple variables $x, y \in X$

the degree of a polynomial is the highest exponentiation in a polynomial:

$$degree(3x^2 + x + 4) = 2$$

the leading coefficient is the coefficient with the highest degree:

$$lco(3x^2 + x + 4) = 3$$

the constant term is the term where the variable is exponentiated to the $0th$
power:

$$conterm(3x^2 + x + 4) = 4$$

polynomial interpolation constructs an n-degree polynomial from $n+1$ points;
lagrange interpolation finds the lowest degree polynomial that interpolates a
given set of data

linear polynomials do not require exponentiation and consist solely of the form:

$$a, b \in R, x \in X : a + bx$$

#### Polynomial Algebra

as mentioned above, polynomial rings form, well a ring;

the addition of polynomials is defined by adding the coefficients of variables
with equivalent exponents.

$$
\begin{gather*}
(a_0 x^0 + a_1 x^1 + .. + a_n x^n) = a \in R[X] \\
(b_0 x^0 + b_1 x^1 + .. + b_n x^n) = b \in R[X] \\
((a_0 + b_0)x^0 + (a_1 + b_1)x^1 + .. + (a_n + b_n)x^n) = c \in R[X] \\
a + b = c
\end{gather*}
$$

the multiplication of polynomials is defined by multiplying each polynomial term
with one another. simplification through term addition is also possible.

$$
\begin{gather*}
(a_0 x^0 + a_1 x^1 + .. + a_n x^n) = a \in R[X] \\
(b_0 x^0 + b_1 x^1 + .. + b_n x^n) = b \in R[X] \\
((a_0 b_0)(x^0 x^0) + (a_0 b_1)(x^0 x^1) + .. + (a_n b_n)(x^n x^n)) = c \in R[X] \\
a * b = c
\end{gather*}
$$

reiterating for clarity, these addition and multiplication operations over
polynomials form a ring, which exhibits the following algebraic properties:

$$
\begin{gather*}
    \forall a, b \in R[X] : a + b = c \in R[X] \\
    \forall a, b, c \in R[X] : (a + b) + c = a + (b + c) \\
    \forall a \in R[X], \exists e \in R[X] : a + e = e + a = a \\
    \forall a \in R[X], \exists a^{-1} \in R[X] : a + a^{-1} = a^{-1} + a = e \\
    \forall a, b \in R[X] : a + b = b + a \\
    \\
    \forall a, b, c \in R[X] : (a * b) * c = a * (b * c) \\
    \forall a \in R[X], \exists e \in R[X] : a * e = e * a = a \\
    \\
    \forall a, b, c \in R[X] : a * (b + c) = (a * b) + (a * c)
\end{gather*}
$$

in addition to polynomial addition, subtraction, and multiplication, there
exists scalar polynomial multiplication, which multiplies a scalar
against each coefficient in the polynomial

$$a, b \in R, x \in X : (a_0 x^0 + a_1 x^1 + .. a_n x^n) \times b$$

## Morphisms

### Ring Homomorphism

a ring homomorphism is a function between two rings that preserves addition,
multiplication, and multiplicative identity:

$$f : R \rightarrow S$$

$$f(a + b) = f(a) + f(b)$$

$$f(ab) = f(a)f(b)$$

$$f(1_R) = 1_S$$

### Ring Isomorphism

if $f : R \rightarrow S$ is a bijection, that is to say every element of $S$ is
mapped from exactly one element each from $R$, then the inverse of the function
$f^{-1}$ is also a ring homomorphism, which forms a ring isomorphism
